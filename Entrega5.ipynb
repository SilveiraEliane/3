```python
#data frame Dados
data=pd.DataFrame(dados_problema)
data.head(20)
```

```python
#soma Demanda *Custo Regional COG
somaproduto = np.sum(data['Demanda (Paletes)']*data['Custo Unitário (/pallet/km)'])
```

```python
#Criando o TUPLE de coordenadas para calcular COG
data['coordinates'] = data.apply(lambda row: (row['Latitude'], row['Longitude']), axis=1)
```

```python
data.head(15)
```

```python
#Cálculo Coordenada
data['lat_ponderada'] = data.apply(lambda row: row['Demanda (Paletes)'] * row['Custo Unitário (/pallet/km)'] * row['coordinates'][0], axis=1)
lat_cog = round(data['lat_ponderada'].sum() / somaproduto, 4)
```

```python
data['long_ponderada'] = data.apply(lambda row: row['Demanda (Paletes)'] * row['Custo Unitário (/pallet/km)'] * row['coordinates'][1], axis=1)
long_cog = round(data['long_ponderada'].sum() / somaproduto, 4)
```

```python
#Resultado Coordenada LAT, LONG do CD
print('O CD deve ser localizado próxio das coordenadas:{},{}'.format(lat_cog, long_cog))
```

```python
!pip install geopy
from geopy.distance import geodesic
```

```python
#Cálculo das Distâncias
for index, row in data.iterrows():
  coordenadas_centro = (lat_cog, long_cog)
  coordenadas_locais = (row['Latitude'], row['Longitude'])
  distance = round(geodesic(coordenadas_centro, coordenadas_locais).kilometers, 2)
  print(f'Distância do centro para {row["Local"]}: {distance} km')
```

```python
#Cálculo da Distância da Rua
def calculate_road_distance(origin, destination):
    url = "http://dev.virtualearth.net/REST/v1/Routes/DistanceMatrix"
    params = {
        "origins": f"{origin[0]}, {origin[1]}",
        "destinations": f"{destination[0]}, {destination[1]}",
        "travelMode": "driving",
        "key": "ApDf65gx4OAKxl8mq7WVRNG1KVYvIuTYQ7V_TC6sIYRzoML94Hyu-GYhxOHLEyDz"
    }
    response = requests.get(url, params=params)
    data = response.json()
    if "resourceSets" in data and data["resourceSets"]:
        results = data["resourceSets"][0]["resources"]
        if results:
           return results[0]["results"][0]["travelDistance"]
    return None


for index, row in data.iterrows():
    location_coordinates = (row['Latitude'], row['Longitude'])
    distance = calculate_road_distance((lat_cog, long_cog), location_coordinates)
    if distance is not None:
        print(f'Distância da rua para {row["Local"]}: {round(distance, 3)} km')
    else:
        print(f'Não foi possível calcular a distância da rua para {row["Local"]}')
```
